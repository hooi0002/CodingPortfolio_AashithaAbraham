```python
# Immutable Python Objects
# Integer, String, Tuples, Floats, Boolean, Complex

x = 100
print(id(x))

x = 50
print(id(x))
```

```python
y = 10
print(id(y))

y += 1
print(id(y))
```

```python
x = 'NDP2023'
print(id(x))

x = 'NDP2024'
print(id(x))
```

```python
def func1(testString):
  print(id(testString))


str1 =  "Taiwan"
print(id(str1))

func1(str1)
# A memory loc can have multiple references
```

```python
# Mutable Python Objects
# Lists, Sets, Dictionaries

foodList = ["Pizza", "HDL", "BBT"]
print(id(foodList))

# To edit an element in the list
foodList[0] = "Durian"
print(foodList)
print(id(foodList))
```

```python
# To edit multiple items in the list
numList1 = [1, 2, 3, 4, 5, 6]
print(numList1)
print(id(numList1))

numList1[1:5:1] = [0, 0, 0, 0]
print(numList1)
print(id(numList1))
```

```python
# Another example
numList1 = [1, 2, 3]
print(id(numList1))

numList1.append(4)
print(id(numList1))
```

```python
numList1 = [100, 100, 100]
print(id(numList1))

# Re-assignment
numList1 = [200, 200, 200]
print(id(numList1))
```

```python
# Demonstrates the mutability of a list
numList1 = [100, 100, 100]
print(id(numList1))

numList2 = numList1
numList2[0] = 0
print(numList2)

print(numList1)
```

```python
numList1 = [100, 100, 100]
print(id(numList1))

numList2 = numList1.copy()
numList2[0] = 0
print(numList2)
print(id(numList2))

print(numList1)
```

```python
# Tuples - Immutable Python Objects
tup1 = (1, 2, 3)
print(id(tup1))

tup1 = tup1 + (4,)
print(tup1)
print(id(tup1))
```

```python
list1 = [1, 2, 3]
print(id(list1))

list1 = list1 + [4] # Reassignment
print(list1)
print(id(list1))
```

```python
list1 = [1, 2, 3]
print(id(list1))

list1.append(4)
print(list1)
print(id(list1))
```
