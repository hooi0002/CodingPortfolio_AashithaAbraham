```python
# Scope of programming in Python

"""
Common abbreviation: LEGB

Local - Variables that are defined within a function
Enclosing - Variables that are defined in the local scope of an enclosing function
Global - Variables defined at the top level of a module/script or those
          declared explicitly with the "global" keyword
Built-in - Variables that are pre-assigned in Python
"""
```

```python
# Global scope

str1 = 'Python'

print(str1)
```

```python
# Global scope
bank = 'DBS BANK'

def f1():
  # Local scope
  # bank = 'DBS BANK'
  print(bank)

f1()
```

```python
# Global scope
place = 'Clarke Quay'

def parent():
  # Enclosing scope
  # place = 'Punggol'

  def child():
    # Local scope
    # place = 'Jurong'
    print(place)

  child()

parent()
```

```python
# Variable shadowing
# Using the same variable name in different scopes

# Global scope
x = 10

def outer_func():
  # Enclosing scope
  x = 20

  def inner_func():
    # Enclosing scope
    # x = 30
    nonlocal x

    def inner_inner_func():
      # Local scope
      # x = 40
      # global x
      nonlocal x
      print(x)

    inner_inner_func()
  inner_func()
outer_func()
```

```python
def f2(var):
  # Local scope
  print('Do I know that variable?', var)
  var += 1
  print(var)

# Global scope
num = 1

f2(num)

print(num)
```

```python
# Namespace modification

# Global scope
x = 10

def modify_global():
  # Local scope
  global x
  x = 30


def modify_local():
  x = 20


modify_global()
print(x)
```

```python
def swap(x, y):
  # Local scope
  z = x
  x = y
  y = z
  print("Swaped in the local scope:", x, y)


x = 5
y = 10
swap(x, y)

print("Global x, y:", x, y)
```

```python
def swap():
  # Local scope
  global x
  global y
  z = x
  x = y
  y = z
  print("Swaped in the local scope:", x, y)


x = 5
y = 10
swap()

print("Global x, y:", x, y)
```
