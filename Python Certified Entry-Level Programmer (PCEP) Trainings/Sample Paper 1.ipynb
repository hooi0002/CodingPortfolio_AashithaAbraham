```python
# for-else
# while-else

# The else-case is literally a no-break case
# If there is no break statement hit in the loop body, the else clause will be executed

greeting = "Good Morning"

for ch in greeting:
  if ch == 'o':
    break
  print(ch)

else:
  print("Good Night")
```

```python
for i in range(0, 5, 1):
  if i == 10:
    break
else:
  print("No break hit")
```

```python
# The break statement will ONLY terminate the innermost loop it is placed within
for i in range(0, 5, 1):
  for j in range(0, 5, 1):
    if j == 0:
      break
  print("Outer loop is still running")
```
