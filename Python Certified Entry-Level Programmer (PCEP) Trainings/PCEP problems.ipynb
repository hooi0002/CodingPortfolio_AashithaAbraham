```python
"""
In Python, the break statement will only terminate the innermost loop it is placed within.
When encountered, the break statement immediately exits the loop it is inside and continues
with the next statement after the loop.

If you have nested loops (i.e., an inner loop within an outer loop),
and you place the break statement in the inner loop, it will only break out of that inner loop.
The outer loop will continue executing until its own termination condition is met.

Here's an example to illustrate this:
"""

for i in range(3):
    print("Outer loop:", i)
    for j in range(3):
        print("Inner loop:", j)
        if j == 1:
            break

"""
In this example, the break statement is placed inside the inner loop with the condition if j == 1.
As you can see, when j is equal to 1, the inner loop is terminated,
but the outer loop continues its execution until it completes all iterations.

If you want to break out of both the inner and outer loops simultaneously,
you can use a flag variable or consider using other control flow structures like return, raise,
or sys.exit() depending on the context of your code.

"""
```

```python
# How do I break out of both the inner and outer loops simultaneously?

flag = False

for i in range(3):
    print("Outer loop:", i)
    for j in range(3):
        print("Inner loop:", j)
        if j == 1:
            flag = True
            break
    if flag:
        break
```
