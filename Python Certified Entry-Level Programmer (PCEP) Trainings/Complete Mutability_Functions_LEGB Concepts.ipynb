```python
## Argument passing
# Integer - Immutable
x = 30

def Change_me(anyNum):
  anyNum = 0


Change_me(x)

print(x)
```

```python
## Argument passing
# String - Immutable
y = 'BlackPink'

def Change_me(anyStr):
  anyStr = 'BlackBlack'


Change_me(y)

print(y)
```

```python
## Argument passing
# List ITEM - Mutable
z = ['A', 'B', 'C']

def Change_me(anyList):
  anyList[0] = 'D'


Change_me(z)

print(z)
```

```python
## Argument passing
# List Item - Mutable
b = ['A', 'B', 'C']

def Change_me(anyList):
  anyList.append('D')


Change_me(b)

print(b)
```

```python
## Argument passing
# List ressignment - Immutable
a = ['A', 'B', 'C']

def Change_me(anyList):
  # Re-assign a new list
  anyList = ['X', 'Y', 'Z']


Change_me(a)

print(a)
```

```python
## Argument passing
# List reassignment - Imutable
c = ['A', 'B', 'C']

def Change_me(anyList):
  anyList = []


Change_me(c)

print(c)
```

```python
# Accessing (Read) a variable in Python
# When NO INPUT is provided in the function!
# Obey LEGB rule

# Global scope
x1 = 100

def func():
  # Local scope
  print(x1)


func()
```

```python
# Accessing (Modify) a variable in Python
# Obey LEGB rule

# Global scope
x1 = 100
print(id(x1))

def func():
  # Local scope
  global x1
  x1 += 1
  """
  # Integer is immutable,
  but since there's no new variable involved (no argumnet is passed into the function),
  Python assigned 101 in a different memory location that has the same variable x1
  which is why it seems like the original x1 is affected.
  the two x1 address are different!
  """
  print(x1)


func()
print(id(x1))
print("Original x1 value:", x1)
```
