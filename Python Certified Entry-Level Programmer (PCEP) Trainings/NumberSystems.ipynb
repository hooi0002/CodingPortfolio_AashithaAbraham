```python
# 1 byte = 8 bits

bin1 = 0b1000
oct1 = 0o20
hex1 = 0xFF

# Convert binary to denary (base 10)
print(int('0b1000', 2))
print(int('0o20', 8))
print(int('0xFF', 16))
```

```python
int1 = 100

# Denary to binary
print(bin(int1))

# Denary to oct
print(oct(int1))

# Denary to hex
print(hex(int1))
```

```python
# Math operation of binary numbers
bin1 = 0b111
bin2 = 0b11

print(bin1, bin2)
print(bin(bin1 + bin2))
```

```python
# Bitwise Operators
# It is to manipulate a data piece on its granular(bit) level

# BITWISE AND, &
# Both inputs have to be 1 for the output to be 1

x1 = 0b100
x2 = 0b111
print(x1 & x2)
print(bin(x1 & x2))

print(bin(0b0 & 0b0))


1 > 3 and 1 > 5
```

```python
# BITWISE OR, |
# Either one of the input has to be 1 for the output to be 1

x1 = 0b1101
x2 = 0b100

print(bin(x1 | x2))
```

```python
# BITWISE XOR, ^
# Both of the inputs have to be different for the output to be 1
x1 = 0b1101
x2 = 0b1001

print(bin(x1 ^ x2))
```

```python
# BIT SHIFTING
# left shifting, << (multiply the number by 2 for each left shift)
x1 = 0b100
print(x1)

print(bin(x1 << 3))
print(x1 << 3)
```

```python
# BIT SHIFTING
# right shifting, >> (divide the number by 2 for each right shift)

x1 = 0b101010
print(bin(x1 >> 1), (x1 >> 1), sep=' - ')
print(bin(x1 >> 2), (x1 >> 2), sep= ' - ')
print(bin(x1 >> 3), (x1 >> 3), sep=' - ')
```

```python
def bitwise_addition(a, b):
    while b != 0:
        carry = a & b
        a = a ^ b
        b = carry << 1
    return a

# Test
result = bitwise_addition(5, 3)
print(result)  # Output: 8
```

```python
# Write a program to toggle the nth bit of a given integer
# (1 << n-1) is the mask
def toggle_bit(num, n):
  return num ^ (1 << n-1)

toggle_bit(10, 3)
```

```python
# Write a program to set the nth bit of a given integer
def set_bit(num, n):
  return num | (1 << n - 1)

set_bit(10, 3)
```

```python
# a program to clear the nth bit of a given integer
def clear_bit(num, n):
  return num & ~(1 << n - 1)

"""
  1010
& 1101
------
  1000
"""

clear_bit(10, 2)
```

```python
# Write a program to check if the nth bit of a given integer is set or not
# Function returns Boolean True/False
# != NOT EQUAL
# In this case, the mask ia a binary data which has all zeros except for the nth bit
def check_bit(num, n):
  return (num & (1 << n-1)) != 0

# Analogy
# num = face
# & = how do we apply the mask
# (1 << n - 1) = mask type

check_bit(12, 4)
```

```python
def toggle_first_last_bit(num):
  print(bin(num))
  msb = num.bit_length()
  mask = (1 << msb - 1)
  num ^= mask # num = num ^ mask, toggle first bit

  num ^= 1 # toggle last bit
  return bin(num)

print(toggle_first_last_bit(22))
```

```python
def insert_one_bit(num, bit, i):
  print(bin(num))

  left = num >> i
  left = (left << 1) | bit
  left = left << i

  right = ((1 << i) - 1) & num # a formula to extract bits up to position i from lsb

  return bin(right | left)
  

print(insert_one_bit(0b1011, 1, 2))


# AND, & - To find differences
# OR, | - To combine
# XOR, ^ - To toggle
# <<, to add 0 bits behind, for alignment
# >>, to trim trailing bits, for alignment
```

```python
def remove_bit(num, i):
  left = num >> (i + 1)
  left = left << i

  right = ((1 << i) - 1) & num
  return left | right

remove_bit(0b1011, 2)
```
